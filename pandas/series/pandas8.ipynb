```python
import pandas as pd
```

```python
s = pd.Series([10,20,30,0,None,40,None,50])
print(s)
print(s.isnull())
print(s.isnull().sum())
print(s.notnull())
print(s.notnull().sum())
```

```python
s = s.fillna(0)# Filling null and NaN values with 0
print(s)
print(s.isnull().sum())
```

```python
s = pd.Series([10,20,30,0,None,40,None,50])
print(s)
s = s.dropna()# Dropping null and NaN values
print(s)
```
