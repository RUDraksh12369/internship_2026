```python
import pandas as pd
```

```python
s2 = pd.Series({"Maths":90,"Science":85,"English":92,"Hindi":88,"Physics":78,"Chemistry":82,"Computer":95},dtype=float)
```

```python
print(s2.head())#shows first 5 values
print(s2.head(3))#shows first 3 values
print(s2.tail())#shows last 5 values
print(s2.tail(2))#shows last 2 values
```

```python
print(s2.sample())#shows random value
print(s2.shape)#shows shape
print(s2.size)#shows size
print(s2.ndim)#shows no ofdimension
print(s2.dtype)#shows datatype
print(s2.values)#shows values
print(s2.index)#shows index
print(s2.info())#shows info
```
