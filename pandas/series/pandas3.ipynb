```python
import pandas as pd 
```

```python
s = pd.Series([1, 2, 3, 4, 5], index=['a','b','c','d','e'])
print(s)
print(s['a'])
print(s[0])
```

```python
print(s.loc['c'])# Access by label
print(s.iloc[2])# Access by index position
```

```python
print(s.at['c'])# Access by label using at
print(s.iat[2])# Access by index position using iat
```

label based slicing

```python
print(s[['a','c']])
print(s['a':'c'])# Slicing by label
```

boolean access

```python
print(s>3)
```
