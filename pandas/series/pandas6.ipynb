```python
import pandas as pd
```

```python
match1 = pd.Series([100,200,300,400,500],index=["a","b","c","d","e"])
match2 = pd.Series([100,200,300,400,500],index=["x","b","c","d","e"])
```

```python
result = match1.add(match2,fill_value=0)# Addition with alignment and fill_value
print(result)
```

```python
result = match1.sub(match2,fill_value=0)# Subtraction with alignment and fill_value
print(result)
```

```python
result = match1.mul(match2,fill_value=1)# Multiplication with alignment and fill_value
print(result)
```

```python
result = match1.div(match2,fill_value=1)# Division with alignment and fill_value
print(result)
```
