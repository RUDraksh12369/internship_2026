```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"Fruit":["apple","banana","orange","grape"],"vegitables":["potato","tomato","ladyfinger","carrot"],"Market_Price":[100,40,80,60]})
df2 = pd.DataFrame({"Fruit":["apple","banana","grape","kiwi"],"vegitables":["potato","tomato","cabbage","carrot"],"Shop_Price":[120,50,90,70]})
```

```python
x = pd.merge(df1,df2,on=["Fruit","vegitables"],how="inner")#inner join will return only the common rows between the two dataframes based on the "Fruit" and "vegitables" columns.
print(x)
```

```python
x = pd.merge(df1,df2,on ="vegitables",how="inner")#inner join will return only the common rows between the two dataframes based on the "vegitables" column.
print(x)
```
