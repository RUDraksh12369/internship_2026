```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Movieratingdata1.csv")
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
print(df)
```

```python
df.dropna(inplace=True)#drop all rows with NaN values
print(df)
```

```python
df.dropna(inplace=True,axis=1)#drop all columns with NaN values
print(df)
```

```python
df.dropna(inplace=True,how="any")#drop all rows with any NaN values
print(df)
```

```python
df.dropna(inplace=True,how="all")#drop all rows with all NaN values
print(df)
```
