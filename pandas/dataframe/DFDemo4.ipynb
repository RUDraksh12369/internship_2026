```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name":["raj","amit","sumit","kunal","Neha","Sumit","Sandhya"],
    "Age":[21,22,23,24,25,26,27],
    "City":["Delhi","Mumbai","Bangalore","Delhi","Mumbai","Bangalore","Delhi"]
}
df = pd.DataFrame(data)
```

```python
df["Age"] = df["Age"] + 1
df["Age"] = [22,23,24,25,26,27,28]
print(df)
```

```python
current_year = 2026
df["Born Year"] = current_year - df["Age"]
print(df)

```

```python
df = df.rename(columns={"Age": "Student Age"})
print(df)
```

```python
df.loc[len(df)] = {
    "Name": "Rohit",
    "Student Age": 28,
    "City": "Pune",
    "Marks": 89,
    "Born Year": 2026 - 28
}
print(df)
```

```python
df["Marks"] = [85, 78, 90, 88, 92, 80, 87, 89]
print(df)
```
