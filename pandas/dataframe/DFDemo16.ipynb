
1. movie , director  ==> join  
2. drop   ==> extra col / row  
3. find  top 10  director  ==> movie count 



```python
import  pandas as pd 
```

```python
movies = pd.read_csv("movies.csv")
directors = pd.read_csv("directors.csv")

movies =movies.drop(columns="Unnamed: 0")
directors=directors.drop(columns="Unnamed: 0")
```

```python
df=pd.merge(
    movies,
    directors,
    left_on="director_id", 
    right_on="id",
    how="left"
)
```

```python
print(df.head())
print(movies.head())
print(directors.head())
```

```python
top_10_directors = df.groupby("director_name")["title"].count().sort_values(ascending=False).head(10)
print(top_10_directors)
```
