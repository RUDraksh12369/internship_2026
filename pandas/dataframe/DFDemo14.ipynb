```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"Fruit":["apple","banana","orange","grape"],"Market_Price":[100,40,80,60]})
df2 = pd.DataFrame({"Fruit":["apple","banana","grape","kiwi"],"Shop_Price":[120,50,90,70]})
```

```python
x = pd.merge(df1, df2, on="Fruit", how="inner")#inner join will return only the common rows between the two dataframes based on the "Fruit" column.
print(x)
```

```python
x = pd.merge(df1,df2,on="Fruit",how="outer")#outer join will return all the rows from both dataframes, and fill in NaN for missing values where there is no match.
print(x)
```

```python
x = pd.merge(df1,df2,on="Fruit",how = "left")#left join will return all the rows from the left dataframe (df1) and the matching rows from the right dataframe (df2). If there is no match, it will fill in NaN for the columns from df2.
print(x)
```

```python
x = pd.merge(df1,df2,on="Fruit",how = "right")#right join will return all the rows from the right dataframe (df2) and the matching rows from the left dataframe (df1). If there is no match, it will fill in NaN for the columns from df1.
print(x)
```
