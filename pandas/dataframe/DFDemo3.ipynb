```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name":["raj","amit","sumit","kunal","Neha","Sumit","Sandhya"],
    "Age":[21,22,23,24,25,26,27],
    "City":["Delhi","Mumbai","Bangalore","Delhi","Mumbai","Bangalore","Delhi"]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
print(df.loc[2,"Age"])# Accessing single value using .loc where 2 is the index and "Age" is the column name
```

```python
print(df.iloc[2,1])# Accessing single value using .iloc where 2 is the index and 1 is the column position
```

```python
print(df["Age"])# Accessing entire column 'Age'
```

```python
print(df[["Age","City"]])# Accessing multiple columns 'Age' and 'City'
```

```python
print(df.iloc[0])# Accessing entire first row using .iloc
```
