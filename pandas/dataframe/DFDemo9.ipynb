```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Movieratingdata1.csv")
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
```

```python
df.replace(["not available","not mension","not value"],np.nan,inplace=True)#replace specific values with NaN
df.fillna(value=0,inplace=True)#replace all NaN values with 0
df.fillna({"MOVIE_ID":0},inplace=True)#replace NaN values in MOVIE_ID column with 0
df.fillna({"MOVIE_ID":0,"PRODUCTION_YEAR":2000},inplace=True)#replace NaN values in MOVIE_ID column with 0 and PRODUCTION_YEAR column with 2000
```
