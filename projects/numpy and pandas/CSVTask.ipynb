```python
# ans 1
import pandas as pd
import numpy as np

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

df = pd.read_csv("Movieratingdata1.csv")
df
```

```python
# ans 2
df = pd.read_csv(
    "Movieratingdata1.csv",
    na_values={"PRODUCTION_YEAR": ["not available", "not mension"]}
)
df
```

```python
# ans 3
df.replace(
    ["not available", "not mension", "not value"],
    np.nan,
    inplace=True
)
df
```

```python
# ans 4
df.isna().sum().sum()
```

```python
# ans 5
df.isna().sum()
```

```python
# ans 6
df.dropna(how="all", inplace=True)
df
```

```python
# ans 7
df_no_nan_cols = df.dropna(axis=1, how="any")
df_no_nan_cols
```

```python
# ans 8
df["MOVIE_ID"] = df["MOVIE_ID"].fillna(0)
df
```

```python
# ans 9
df["PRODUCTION_YEAR"] = df["PRODUCTION_YEAR"].fillna(2000)
df
```

```python
# ans 10
df["RATING"] = df["RATING"].astype(float)
df["RATING"] = df["RATING"].fillna(df["RATING"].mean())
df
```

```python
# ans 11
df.replace("1390411", "The Batman", inplace=True)
df
```

```python
# ans 12
df.columns = df.columns.str.lower()
df
```

```python
# ans 13
df["production_year"] = df["production_year"].astype(int)
df
```

```python
# ans 14
df.drop_duplicates(inplace=True)
df
```

```python
# ans 15
df.dtypes
```

```python
# ans 16
df.reset_index(drop=True, inplace=True)
df
```

```python
# ans 17
df.drop(columns=["movie_title"], inplace=True)
df
```

```python
# ans 18
df.fillna("Unknown", inplace=True)
df
```
