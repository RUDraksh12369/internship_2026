```python
import numpy as np
```

```python
a = np.array([1, 2, 3,4,5,6])
print(a)
print(a.shape)
```

```python
a1 = a.reshape(2,3)
print(a1)
print(a1.shape)
```

```python
a2 = a.reshape(3,2)
print(a2)
print(a2.shape)
```

```python
x = np.arange(12)
b = x.reshape(3,4)
print(b)
```

```python
x2 = np.arange(12)
b2 = x2.reshape(-1,3)
print(b2)
```
