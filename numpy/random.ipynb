```python
import random as r
import numpy as np
```

```python
print(r.random())# number between 0 and 1 1 excluded
```

```python
print(r.randrange(1,10,2))# odd numbers between 1 and 10 taking step 2 excluding 10
```

```python
print(r.randint(1,10))# number between 1 and 10 both included
```

```python
a = np.random.randint(-10,10,12).reshape(3,4)#random values between -10 and 10 in 3x4 matrix
print(a)
```

```python
b = np.random.random((3,4))# random values between 0 and 1 in 3x4 matrix
print(b)
```

```python
a = np.array([[10,2,34],[45,67,89],[15,17,99]])
print(a)
print(a.min())
print(a.max())
```

```python
print(a.argmin())#index of min value
print(a.argmax())#index of max value
```

```python
print(a.mean())
print(np.median(a))
print(a.std())
```

```python
print(a.argmax(axis=0))#index of max value column wise
print(a.argmax(axis=1))#index of max value row wise
print(a.argmin(axis=0))#index of min value column wise
print(a.argmin(axis=1))#index of min value row wise
```

```python
print(np.sort(a))# sort each row
print(np.argsort(a,axis=1))# indices that would sort each row
print(np.sort(a, axis=0))# sort each column
print(np.argsort(a,axis=0))# indices that would sort each column
```

```python
a = np.arange(0,9).reshape((3,3))
b = np.arange(9,18).reshape((3,3))
result = np.dot(a,b)
print(result)
result = a.dot(b)
print(result)
result = np.matmul(a,b)
print(result)
```
