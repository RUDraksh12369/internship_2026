```python
#ans 9
arr3d = np.arange(1, 28).reshape(3,3,3)

print("3D Array:\n", arr3d)

# Overall mean
overall_mean = np.mean(arr3d)

# Mean along axis 0
axis0_mean = np.mean(arr3d, axis=0)

# Median of entire array
median_all = np.median(arr3d)

print("\nOverall Mean:", overall_mean)
print("\nMean along axis 0:\n", axis0_mean)
print("\nMedian of entire array:", median_all)
```

```python
# ans 10
arr = np.arange(1, 17)

arr2 = arr.reshape(4,4)
print( arr2)

mult = arr2 * np.array([1,2,3,4])
print("\nAfter column multiplication:\n", mult)

flat = mult.flatten()
print("\nFlattened array:\n", flat)

std_val = np.std(flat)
print("\nStandard Deviation:", std_val)
```
