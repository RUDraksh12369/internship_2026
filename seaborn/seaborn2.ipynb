```python
import seaborn as sns
import pandas as pd
```

```python
sns.set_theme()
sns.set_style("whitegrid")   # whitegrid, darkgrid, dark, white, ticks
sns.set_context("notebook")  # paper, notebook, talk, poster
```

```python
data = {
    "Day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "Sales": [120, 150, 170, 160, 180, 200, 190],
    "Profit": [30, 45, 50, 48, 60, 70, 65],
    "Customers": [20, 25, 30, 28, 35, 40, 38],
    "Category": ["A", "A", "B", "B", "A", "B", "A"]
}

df = pd.DataFrame(data)
```

```python
sns.lineplot(data=df, x="Day", y="Sales")
```

```python
sns.barplot(data=df, x="Day", y="Profit")
```

```python
numeric_df = df[["Sales", "Profit", "Customers"]]
sns.heatmap(numeric_df.corr(), annot=True)
```

```python
sns.boxplot(data=df, x="Category", y="Sales")
```

```python
sns.countplot(data=df, x="Category")
```

```python
sns.scatterplot(data=df, x="Customers", y="Sales", hue="Category")
```

```python
sns.pairplot(df, hue="Category")
```

```python
sns.violinplot(data=df, x="Category", y="Profit")
```
