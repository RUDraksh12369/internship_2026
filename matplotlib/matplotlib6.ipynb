2x2 line,bar,scatter,histogram, all have same data 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
boys_marks  = [35,40,50,65,70,75]
girsl_marks = [36,42,55,70,75,80]
```

```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(boys_marks, girsl_marks)
ax[0, 0].set_title("Line Plot")

ax[0, 1].bar(boys_marks, girsl_marks)
ax[0, 1].set_title("bar Plot")


ax[1, 0].scatter(boys_marks, girsl_marks)
ax[1, 0].set_title("scatter Plot")

ax[1, 1].boxplot(boys_marks, girsl_marks)
ax[1, 1].set_title("box Plot")
```
