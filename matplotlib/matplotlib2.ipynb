```python
student = ["Raj","Parth","Amit","Sumit","Jay"]
marks = [80,78,56,71,69]
colors = ["red","green","Blue","yellow","Black"]
plt.bar(student,marks,color = colors)
plt.title("Student Marks")
plt.xlabel("Name")
plt.ylabel("Marks")
for i in range(len(student)):
    plt.text(i,marks[i],str(marks[i]),ha="center",va="bottom")
plt.show()
```

```python
order_timing=[11,12,13,14,15]
punjabi_order = [20,43,56,32,12]
italian_order = [34,21,45,32,45]
chainese_order = [23,56,43,23,67]
```

```python
plt.figure(figsize=(11, 7))
plt.bar(order_timing,punjabi_order,label="Punjabi orders")
plt.bar(order_timing,italian_order,bottom=punjabi_order,label="italian_order")  
plt.bar(order_timing,chainese_order,bottom=[punjabi_order[i]+italian_order[i] for i in range(len(order_timing))],label="chainese_order")
plt.legend()
```

```python
subjects=["Maths","Science","Eng"]
boys = [70,80,75]
girls=[75,85,70]
```

```python
x = np.arange(len(subjects))
width=0.35
```

```python
plt.bar(x-width/2,boys,width,label="Boys")
plt.bar(x+width/2,girls,width,label="Girls")
plt.xticks(x,subjects)
plt.legend()
```
